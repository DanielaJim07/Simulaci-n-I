```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math

def inversa(n):
  x = []
  for i in range(n):
    R = random()
    if 0<=R<= 0.5:
      x. append(math.sqrt(2*R))
    else:
      x.append(2*R)
  return x

a = inversa(100000)

plt.hist(a, bins=50)

print(np.mean(a))
print(np.var(a))
```
