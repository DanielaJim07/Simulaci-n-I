```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate


a = 1001
c = 421
m = 100**2
x = 5  # x0

valores = [["n", "x_n", "x_n / m"]]
muestra = []

for n in range(10000):
    valores.append([n, x, x / m])
    muestra.append(x / m)
    x = (a * x + c) % m

print(tabulate(valores[:20], headers="firstrow", tablefmt="fancy_grid", numalign="center"))

media = np.cumsum(muestra) / np.arange(1, len(muestra) + 1)
varianza = np.array([np.var(muestra[:i+1]) for i in range(len(muestra))])
acumulada = np.cumsum(muestra)


plt.figure(figsize=(10, 5))
plt.plot(media, label="Media", linestyle="-", color="c")
plt.plot(varianza, label="Varianza", linestyle="-", color="m")
plt.plot(acumulada, label="Acumulada", linestyle="--", color="orange")

plt.legend()
plt.xlabel("n")
plt.ylabel("Valor")
plt.title(" Media, Varianza y Suma Acumulada")
plt.grid(color="lightblue", linestyle="-", linewidth=0.7)
plt.show()
```
