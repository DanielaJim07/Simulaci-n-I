```python
import numpy as np

# Definimos la matriz de transición A
A = np.array([
    [0.3, 0.5, 0.2, 0, 0],
    [8/20, 7/20, 0, 1/20,  4/20],
    [1/5, 1/5,1/5,1/5,1/5],
    [1/3,0, 2/3,  0, 0],
    [1/2,0, 0,  0, 1/2]
])

# Estado inicial:
# Vector: [0,1,2,3,4]
X0 = np.array([1, 0, 0, 0, 0])  # 100% en naranja

# Número de pasos (lanzamientos)
n = 10

# Evolucionamos la cadena de Markov
Xn = np.linalg.matrix_power(A, n) @ X0

# Mostramos el resultado
etiquetas = ['0', '1', '2', '3', '4']
for i, p in enumerate(Xn):
    print(f"Probabilidad de estar en {etiquetas[i]} después de {n} lanzamientos: {p:.4f}")
```

```python
# Utilizamos el valor esperado: E[pacientes] = sum(i * P(i))
promedio_pacientes = np.sum(np.arange(5) * Xn)

print(f"\nNúmero promedio de pacientes en el consultorio después de {n} lanzamientos: {promedio_pacientes:.4f}")
```

```python
import numpy as np

# Definimos la matriz de transición A
A = np.array([
    [0.3, 0.5, 0.2, 0, 0],
    [8/20, 7/20, 0, 1/20,  4/20],
    [1/5, 1/5, 1/5, 1/5, 1/5],
    [1/3, 0, 2/3, 0, 0],
    [1/2, 0, 0, 0, 1/2]
])

# Estado inicial:
X0 = np.array([1, 0, 0, 0, 0])  # 100% en el estado 0 (consultorio vacío)

# Número de pasos (lanzamientos)
n = 10

# Inicializamos el estado
Xn = X0

# Mostramos la probabilidad de estar en cada estado después de 0 pasos (estado inicial)
print(f"Estado inicial (0 lanzamientos): {Xn}")
print()

# Evolucionamos la cadena de Markov paso por paso
for k in range(1, n + 1):
    # Calculamos A^k
    A_k = np.linalg.matrix_power(A, k)

    # Evolucionamos el estado
    Xn = A_k @ X0

    # Imprimimos la matriz de transición A^k
    print(f"Matriz de transición A^{k}:")
    print(A_k)
    print()

    # Imprimimos las probabilidades de estar en cada estado después de k pasos
    etiquetas = ['0', '1', '2', '3', '4']
    print(f"Probabilidades de estar en cada estado después de {k} lanzamientos:")
    for i, p in enumerate(Xn):
        print(f"  Estado {etiquetas[i]}: {p:.4f}")
    print()

# Calcular el número promedio de pacientes en el consultorio (valor esperado)
promedio_pacientes = np.sum(np.arange(5) * Xn)
print(f"Promedio de pacientes en el consultorio después de {n} lanzamientos: {promedio_pacientes:.4f}")
```
