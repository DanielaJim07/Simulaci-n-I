# Método Box-Muller

Generar $$U_1, U_2 \sim U(0,1)$$.

Aplicar :

$$R = \sqrt{-2 \ln U_1}$$  
$$\theta = 2 \pi U_2$$

Finalmente, obtener dos variables aleatorias independientes con distribución normal estándar \(N(0,1)\) mediante:

$$Z_0 = R \cos(\theta) = \sqrt{-2 \ln U_1} \cdot \cos(2 \pi U_2)$$  
$$Z_1 = R \sin(\theta) = \sqrt{-2 \ln U_1} \cdot \sin(2 \pi U_2)$$

Estos valores $$Z_0$$ y $$Z_1$$ tienen distribución normal estándar: $$\mathcal{N}(0, 1)$$


```python
# Importamos las librerías
import random as rd
import numpy as np
import matplotlib.pyplot as plt
from math import log, cos, sin, pi
```

##Método de Box-Muller

    Genera 2n valores que siguen una distribución normal estándar (media 0, varianza 1)
    utilizando el método de Box-Muller.
    
    Parámetros:
    n (int): número de pares de valores a generar (se generarán 2n valores normales).
    
    Retorna:
    list: Lista con 2n valores generados.

```python
# Función que implementa el método de Box-Muller para generar variables normales estándar

def aproximacion(n):
    lista = []

    for i in range(n):
       # Generar dos variables aleatorias uniformes independientes en el intervalo (0,1)
        U1 = rd.random()
        U2 = rd.random()
 # Aplicación de la transformación de Box-Muller
        R = (-2 * log(U1))**0.5
        tt = 2 * pi * U2

        valorx = R * cos(tt)
        lista.append(valorx)

        valory = R * sin(tt)
        lista.append(valory)

    return lista

# Prueba: imprimir una muestra pequeña de números normales generados
print(aproximacion(10))
```

```python
# Graficar el histograma de 100,000 valores generados para visualizar la distribución
plt.hist(aproximacion(100000), bins=50, density=True)
plt.title("Histograma de números normales generados (Box-Muller)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()
```

```python
# Calcular y mostrar la varianza y media
print("Varianza:", np.var(aproximacion(100000)))
print("Media:", np.mean(aproximacion(100000)))
```
