```python
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
a = 81
c = 89
m = 10**2
x = 5  # x0

# Inicializar tabla de valores
valores = [["n", "x_n", "x_n / m"]]

# Generar secuencia
for n in range(100):
    valores.append([n, x, x / m])
    x = (a * x + c) % m

# Imprimir tabla
print(tabulate(valores, headers="firstrow", tablefmt="fancy_grid", numalign="center"))

# Extraer valores
lista_xn_m = [row[2] for row in valores[1:]]  # Omitimos la cabecera
n_values = [row[0] for row in valores[1:]]  # Extraemos los valores de n

# Gráfico
plt.plot(n_values, lista_xn_m, marker="o", linestyle="-", color="c")
plt.title("Gráfico de $x_n / m$", fontsize=14, color="purple")
plt.xlabel("n")
plt.ylabel("$x_n / m$")
plt.grid(color="lightblue", linestyle="-", linewidth=0.7)
ax = plt.gca()
ax.spines["top"].set_color("purple")
ax.spines["bottom"].set_color("purple")
ax.spines["left"].set_color("purple")
ax.spines["right"].set_color("purple")


for spine in ax.spines.values():
    spine.set_linewidth(0.5)
plt.show()

# Histograma
plt.hist(lista_xn_m, bins=10, edgecolor="lightblue", color="pink", alpha=0.7)
plt.title("Histograma de $x_n / m$",fontsize=14, color="purple")
plt.xlabel("Valores de $x_n / m$")
plt.ylabel("Frecuencia")
plt.grid(color="lightblue", linestyle="--", linewidth=0.7)
ax = plt.gca()
ax.spines["top"].set_color("purple")
ax.spines["bottom"].set_color("purple")
ax.spines["left"].set_color("purple")
ax.spines["right"].set_color("purple")


for spine in ax.spines.values():
    spine.set_linewidth(0.5)
plt.show()
```
