```python

from tabulate import tabulate
import matplotlib.pyplot as plt
import math
import seaborn as sns
import numpy as np

a = 81
c = 89
m = 10**2
x = 5 # x0


valores = [["n", "x_n", "x_n / m"]]


for n in range(101):
    valores.append([n, x, round(x / m, 4)])
    x = (a * x + c) % m
print(tabulate(valores, headers="firstrow", tablefmt="grid"))

lista = [row[1] for row in valores[3:]]
plt.plot(lista)
plt.title("Grafico")
plt.xlabel("S")
plt.ylabel("valores de x/m")
plt.grind(True)

plt.hist(lista, bins=10)
plt.show()
```
