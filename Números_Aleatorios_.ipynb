```python
from tabulate import tabulate


a = 4
c = 7
m = 22
x = 3  # x0


valores = [["n", "x_n", "x_n / m"]]


for n in range(50):
    valores.append([n, x, round(x / m, 4)])
    x = (a * x + c) % m
print(tabulate(valores, headers="firstrow", tablefmt="grid"))
```
